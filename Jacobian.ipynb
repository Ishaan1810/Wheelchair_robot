```python
T1 = sy.Matrix([[sy.cos(theta1),-sy.sin(theta1)*sy.cos(-m.pi/2),sy.sin(theta1)*sy.sin(-m.pi/2),0*sy.cos(theta1)],[sy.sin(theta1),sy.cos(theta1)*sy.cos(-m.pi/2),-sy.cos(theta1)*sy.sin(-m.pi/2),0*sy.sin(theta1)],[0,sy.sin(-m.pi/2),sy.cos(-m.pi/2),114.3],[0,0,0,1]])

T2 = sy.Matrix([[sy.cos(theta2),-sy.sin(theta2)*sy.cos(m.pi/2),sy.sin(theta2)*sy.sin(m.pi/2),0*sy.cos(theta2)],[sy.sin(theta2),sy.cos(theta2)*sy.cos(m.pi/2),-sy.cos(theta2)*sy.sin(m.pi/2),0*sy.sin(theta2)],[0,sy.sin(m.pi/2),sy.cos(m.pi/2),102.616],[0,0,0,1]])

T3 = sy.Matrix([[sy.cos(0),-sy.sin(0)*sy.cos(0),sy.sin(0)*sy.sin(0),0*sy.cos(0)],[sy.sin(0),sy.cos(0)*sy.cos(0),-sy.cos(0)*sy.sin(0),0*sy.sin(0)],[0,sy.sin(0),sy.cos(0),77.216],[0,0,0,1]])
T4 = sy.Matrix([[sy.cos(theta3+m.pi/2),-sy.sin(theta3+m.pi/2)*sy.cos(0),sy.sin(theta3+m.pi/2)*sy.sin(0),51.8*sy.cos(theta3+m.pi/2)],[sy.sin(theta3+m.pi/2),sy.cos(theta3+m.pi/2)*sy.cos(0),-sy.cos(theta3+m.pi/2)*sy.sin(0),51.8*sy.sin(theta3+m.pi/2)],[0,sy.sin(0),sy.cos(0),0],[0,0,0,1]])

T5 = sy.Matrix([[sy.cos(-m.pi/2),-sy.sin(-m.pi/2)*sy.cos(0),sy.sin(-m.pi/2)*sy.sin(0),0*sy.cos(-m.pi/2)],[sy.sin(-m.pi/2),sy.cos(-m.pi/2)*sy.cos(0),-sy.cos(-m.pi/2)*sy.sin(0),0*sy.sin(-m.pi/2)],[0,sy.sin(0),sy.cos(0),127],[0,0,0,1]])
T6 = sy.Matrix([[sy.cos(0),-sy.sin(0)*sy.cos(-m.pi/2),sy.sin(0)*sy.sin(-m.pi/2),101.6*sy.cos(0)],[sy.sin(0),sy.cos(0)*sy.cos(-m.pi/2),-sy.cos(0)*sy.sin(-m.pi/2),101.6*sy.sin(0)],[0,sy.sin(-m.pi/2),sy.cos(-m.pi/2),0],[0,0,0,1]])
T7 = sy.Matrix([[sy.cos(theta4),-sy.sin(theta4)*sy.cos(-m.pi/2),sy.sin(theta4)*sy.sin(-m.pi/2),0*sy.cos(theta4)],[sy.sin(theta4),sy.cos(theta4)*sy.cos(-m.pi/2),-sy.cos(theta4)*sy.sin(-m.pi/2),0*sy.sin(theta4)],[0,sy.sin(-m.pi/2),sy.cos(-m.pi/2),50.8],[0,0,0,1]])

T8 = sy.Matrix([[sy.cos(0),-sy.sin(0)*sy.cos(0),sy.sin(0)*sy.sin(0),-101.6*sy.cos(0)],[sy.sin(0),sy.cos(0)*sy.cos(0),-sy.cos(0)*sy.sin(0),-101.6*sy.sin(0)],[0,sy.sin(0),sy.cos(0),0],[0,0,0,1]])
T9 = sy.Matrix([[sy.cos(0),-sy.sin(0)*sy.cos(0),sy.sin(0)*sy.sin(0),0*sy.cos(0)],[sy.sin(0),sy.cos(0)*sy.cos(0),-sy.cos(0)*sy.sin(0),0*sy.sin(0)],[0,sy.sin(0),sy.cos(0),-128],[0,0,0,1]])
T10 = sy.Matrix([[sy.cos(theta5),-sy.sin(theta5)*sy.cos(-m.pi/2),sy.sin(theta5)*sy.sin(-m.pi/2),0*sy.cos(theta5)],[sy.sin(theta5),sy.cos(theta5)*sy.cos(-m.pi/2),-sy.cos(theta5)*sy.sin(-m.pi/2),0*sy.sin(theta5)],[0,sy.sin(-m.pi/2),sy.cos(-m.pi/2),-76.2],[0,0,0,1]])
T11 = sy.Matrix([[sy.cos(0),-sy.sin(0)*sy.cos(0),sy.sin(0)*sy.sin(0),0*sy.cos(0)],[sy.sin(0),sy.cos(0)*sy.cos(0),-sy.cos(0)*sy.sin(0),0*sy.sin(0)],[0,sy.sin(0),sy.cos(0),-51.8],[0,0,0,1]])
T12 = sy.Matrix([[sy.cos(theta6),-sy.sin(theta6)*sy.cos(m.pi/2),sy.sin(theta6)*sy.sin(m.pi/2),0*sy.cos(theta6)],[sy.sin(theta6),sy.cos(theta6)*sy.cos(m.pi/2),-sy.cos(theta6)*sy.sin(m.pi/2),0*sy.sin(theta6)],[0,sy.sin(m.pi/2),sy.cos(m.pi/2),-153.42],[0,0,0,1]])
T13 = sy.Matrix([[sy.cos(0),-sy.sin(0)*sy.cos(m.pi),sy.sin(0)*sy.sin(m.pi),102.55*sy.cos(0)],[sy.sin(0),sy.cos(0)*sy.cos(m.pi),-sy.cos(0)*sy.sin(m.pi),102.55*sy.sin(0)],[0,sy.sin(m.pi),sy.cos(m.pi),77.19],[0,0,0,1]])
T14 = sy.Matrix([[sy.cos(theta7),-sy.sin(theta7)*sy.cos(0),sy.sin(theta7)*sy.sin(0),0*sy.cos(theta7)],[sy.sin(theta7),sy.cos(theta7)*sy.cos(0),-sy.cos(theta7)*sy.sin(0),0*sy.sin(theta7)],[0,sy.sin(0),sy.cos(0),-133.5],[0,0,0,1]])
T15 = sy.Matrix([[sy.cos(m.pi/2),-sy.sin(m.pi/2)*sy.cos(m.pi),sy.sin(m.pi/2)*sy.sin(m.pi),66.3*sy.cos(m.pi/2)],[sy.sin(m.pi/2),sy.cos(m.pi/2)*sy.cos(m.pi),-sy.cos(m.pi/2)*sy.sin(m.pi),66.3*sy.sin(m.pi/2)],[0,sy.sin(m.pi),sy.cos(m.pi),0],[0,0,0,1]])
```

```python
M01 = T1
M02 = T1*T2
M03 = M02*T3*T4
M04 = M03*T5*T6*T7
M05 = M04*T8*T9*T10
M06 = M05*T11*T12
M07 = M06*T13*T14*T15
```

```python
#using partial differentiation method (method 2) to find jacobian matrix
#finding position vector
pos_vector = M07[:3,3]

#finding partial differentiation of theta 1 and position matrix
d_t1 = pos_vector.diff(theta1)

#finding partial differentiation of theta 2 and position matrix 
d_t2 = pos_vector.diff(theta2)

#finding partial differentiation of theta 3 and position matrix 
d_t3 = pos_vector.diff(theta3)

#finding partial differentiation of theta 4 and position matrix 
d_t4 = pos_vector.diff(theta4)

#finding partial differentiation of theta 5 and position matrix 
d_t5 = pos_vector.diff(theta5)

#finding partial differentiation of theta 6 and position matrix 
d_t6 = pos_vector.diff(theta6)

#finding partial differentiation of theta 7 and position matrix 
d_t7 = pos_vector.diff(theta7)

#finding z0n for jacobian matrix
#using m 0-n matrices
z_t1 = M01[:3,2]
z_t2 = M02[:3,2]
z_t3 = M03[:3,2]
z_t4 = M04[:3,2]
z_t5 = M05[:3,2]
z_t6 = M06[:3,2]
z_t7 = M07[:3,2]

# finding jacobian matrix
J = sy.Matrix([[d_t1,d_t2,d_t3,d_t4,d_t5,d_t6,d_t7],[z_t1,z_t2,z_t3,z_t4,z_t5,z_t6,z_t7]])
```
